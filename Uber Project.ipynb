# Data Moddeling in Pandas

```python
import io
import requests 
import pandas as pd
```

```python
url = 'https://storage.googleapis.com/uber-raw-data/uber_data.csv'
response = requests.get(url)
df = pd.read_csv(io.StringIO(response.text), sep=',')
df
```

```python
df.info()
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
dupes = df[df.duplicated(keep=False)]
dupes
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df['trip_id'] = df.index
df
```

```python
date_range = pd.date_range(start='2016-01-01', end='2017-01-01', freq='min')[:-1]

# Create DataFrame
dim_datetime = pd.DataFrame(date_range, columns=['DateTime'])
```

```python
dim_datetime = dim_datetime.reset_index(drop=True)
dim_datetime
```

```python
#Creating dim_datetime Dimention

date_range = pd.date_range(start='2016-01-01', end='2017-01-01', freq='T')[:-1]

# Create DataFrame
dim_datetime = pd.DataFrame(date_range, columns=['DateTime'])

dim_datetime = dim_datetime.reset_index(drop=True)

dim_datetime['hour'] = dim_datetime['DateTime'].dt.hour
dim_datetime['day'] = dim_datetime['DateTime'].dt.day
dim_datetime['weekday'] = dim_datetime['DateTime'].dt.weekday
dim_datetime['month'] = dim_datetime['DateTime'].dt.month
dim_datetime['year'] = dim_datetime['DateTime'].dt.year

dim_datetime['datetime_id'] = dim_datetime.index

dim_datetime.rename(columns={'DateTime': 'datetime'}, inplace=True)


dim_datetime = dim_datetime[['datetime_id', 'datetime', 'hour', 'day', 'weekday', 'month', 'year']]

dim_datetime

```

```python
#Creating dim_ratecode Dimention

rate_code_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

dim_ratecode = df[['RatecodeID']].drop_duplicates().reset_index(drop=True)
dim_ratecode['ratecode_id'] = dim_ratecode.index
dim_ratecode['ratecode_name'] = dim_ratecode['RatecodeID'].map(rate_code_type)
dim_ratecode.rename(columns={'RatecodeID':'ratecode_src_id'}, inplace=True)
dim_ratecode = dim_ratecode[['ratecode_id', 'ratecode_src_id', 'ratecode_name']]
dim_ratecode
```

```python
#Creating dim_pickup_location Dimention

dim_pickup_location = df[['pickup_longitude', 'pickup_latitude']].drop_duplicates().reset_index(drop=True)
dim_pickup_location['pickup_location_id'] = dim_pickup_location.index
dim_pickup_location = dim_pickup_location[['pickup_location_id', 'pickup_latitude', 'pickup_longitude']]
dim_pickup_location
```

```python
#Creating dim_dropoff_location Dimention

dim_dropoff_location = df[['dropoff_longitude', 'dropoff_latitude']].drop_duplicates().reset_index(drop=True)
dim_dropoff_location['dropoff_location_id'] = dim_dropoff_location.index
dim_dropoff_location = dim_dropoff_location[['dropoff_location_id', 'dropoff_latitude', 'dropoff_longitude']]
dim_dropoff_location
```

```python
#Creating dim_payment_type Dimention

payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}

dim_payment_type = pd.DataFrame.from_dict(payment_type_name, orient='index', columns=['payment_type_name'])
dim_payment_type['payment_type_id'] = dim_payment_type.index
dim_payment_type = dim_payment_type[['payment_type_id', 'payment_type_name']]
dim_payment_type
```

```python
print(dim_payment_type)
```

```python
df['tpep_pickup_datetime'] = df['tpep_pickup_datetime'].dt.floor('min')
df['tpep_dropoff_datetime'] = df['tpep_dropoff_datetime'].dt.floor('min')
```

```python
dim_payment_type.columns
```

```python
fct_trip = df.merge(dim_datetime, how='left', left_on='tpep_pickup_datetime', right_on='datetime') \
                    .merge(dim_datetime, how='left', left_on='tpep_dropoff_datetime', right_on='datetime', suffixes=('', '_dropoff')) \
                    .merge(dim_ratecode, how='left', left_on='RatecodeID', right_on='ratecode_src_id') \
                    .merge(dim_pickup_location, how='left', left_on=['pickup_longitude', 'pickup_latitude'], right_on=['pickup_longitude', 'pickup_latitude']) \
                    .merge(dim_dropoff_location, how='left', left_on=['dropoff_longitude', 'dropoff_latitude'], right_on=['dropoff_longitude', 'dropoff_latitude']) \
                    .merge(dim_payment_type, how='left', left_on='payment_type', right_on='payment_type_id') \
[['trip_id', 'VendorID', 'datetime_id', 'datetime_id_dropoff', 'passenger_count', 'trip_distance', 'pickup_location_id', 'dropoff_location_id', 'ratecode_id', 'store_and_fwd_flag', 'payment_type_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge', 'total_amount']]
```
